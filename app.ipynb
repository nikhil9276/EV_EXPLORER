```python
# Step 1: Standardize column names
df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")
```

```python
# Step 2: Convert 'cargo_volume_l' to numeric, forcing invalids to NaN
df['cargo_volume_l'] = pd.to_numeric(df['cargo_volume_l'], errors='coerce')
```

```python
# Step 3: Fill missing number_of_cells using group median, then fallback to overall median
df['number_of_cells'] = df.groupby('battery_capacity_kwh')['number_of_cells'].transform(
    lambda x: x.fillna(x.median())
)
```

```python
df['number_of_cells'].isnull().sum()
```

```python
df['number_of_cells'] = df['number_of_cells'].fillna(df['number_of_cells'].median())
```

```python
df['number_of_cells'].isnull().sum()
```

```python
# Step 4: Fill remaining numeric columns using column-wise median (excluding number_of_cells)
numeric_cols = df.select_dtypes(include=['float64', 'int64']).columns.drop('number_of_cells')
df[numeric_cols] = df[numeric_cols].fillna(df[numeric_cols].median())
```

```python
df.info()
```

```python
# Drop any rows that still have NaN values
df = df.dropna().copy()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
# Drop the 'source_url' column as it's not useful for dashboard visuals
df = df.drop(columns=['source_url'])
```

```python
# Rename columns for better readability in Power BI
df = df.rename(columns={
    'brand': 'Brand',
    'model': 'Model',
    'segment': 'Segment',
    'drivetrain': 'Drivetrain',
    'top_speed_kmh': 'Top Speed (km/h)',
    'acceleration_0_100_s': '0–100 km/h Acceleration (s)',
    'efficiency_wh_per_km': 'Efficiency (Wh/km)',
    'range_km': 'Range (km)',
    'battery_capacity_kwh': 'Battery Capacity (kWh)',
    'number_of_cells': 'Number of Battery Cells',
    'fast_charging_power_kw_dc': 'Fast Charging Power (kW DC)',
    'car_body_type': 'Car Body Type',
    'fast_charge_port': 'Fast Charge Port Type',
    'battery_type': 'Battery Type',
    'towing_capacity_kg': 'Towing Capacity (kg)',
    'cargo_volume_l': 'Cargo Volume (L)',
    'length_mm': 'Length (mm)',
    'width_mm': 'Width (mm)',
    'height_mm': 'Height (mm)',
    'powertrain': 'Powertrain',
    'price_inr': 'Price (INR)'
})
```

 # Correlation Heatmap for Numeric Relationships

```python
plt.figure(figsize=(12,8))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

```python
df.columns
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Battery Capacity (kWh)', y='Range (km)',  alpha=0.7)
plt.title("Range vs. Battery Capacity")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Fast Charging Power (kW DC)', y='Range (km)', hue='Battery Capacity (kWh)', alpha=0.7)
plt.title("Charging Speed vs Range")
plt.tight_layout()
plt.show()
```

```python
grouped = df.groupby('Car Body Type')[['Top Speed (km/h)', '0–100 km/h Acceleration (s)']].mean()
grouped.plot(kind='bar', figsize=(10,6))
plt.title("Avg Top Speed & Acceleration by Car Body Type")
plt.ylabel("Value")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
